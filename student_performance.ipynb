```python
# Option 1: Specify the correct path to the file
# Replace this with the actual path where your file is located
# For example: data = pd.read_csv("C:/Users/YourName/Downloads/student-mat.csv")
# Or: data = pd.read_csv("./data/student-mat.csv")

# Option 2: Download the file directly (this will work without having the file locally)
import pandas as pd
from urllib.request import urlretrieve
import zipfile
import os

# Create a temporary directory for our files
os.makedirs('temp', exist_ok=True)

# Download the zip file
zip_path = 'temp/student.zip'
urlretrieve("https://archive.ics.uci.edu/ml/machine-learning-databases/00320/student.zip", zip_path)

# Extract the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('temp')

# Now read the CSV file
data = pd.read_csv('temp/student-mat.csv', sep=';')  # Note: this dataset uses semicolon as separator

print("File downloaded and loaded successfully")
data.head()
```

```python
data.info()
```

```python
# Create pass/fail label
data["pass"] = data["G3"] >= 10
data["pass"] = data["pass"].astype(int)

# Select features
features = ["studytime", "failures", "absences", "G1", "G2"]
X = data[features]
y = data["pass"]

X.head()
```
